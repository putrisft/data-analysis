Coming from Bali, which is a high-traffic tourist destination, and assuming that all tourist destinations have the same quality of attractions, the reality presents data showing that there are several destinations with ratings below average. Therefore, from this issue, it is necessary to further examine destinations that are below average and have the lowest ratings to determine the steps needed to improve the ratings of these tourist destinations using SMART Questions.

# Creating effective questions using the SMART Questions method

1. **Specific**: "Which tourist destination has the lowest rating?"
2. **Measurement**: "How many tourist destinations have ratings below the average?" and "What is the lowest rating among these tourist destinations?"
3. **Action Oriented**: "What actions need to be taken by the local area to improve the rating of that tourist destination?"
4. **Relevant**: "Will these actions lead to an improvement in the rating of the tourist destination?"
5. **Time-Bound**: "For how long will the improvement plan be implemented?"

**Data Cleaning**


*   Removing duplicate data
*   Deleting irrelevant columns (those that don't contain necessary information)

```python
df = df.drop(["Coordinate","Source"], axis=1)
```

```python
df.head()
```

```python
"""
Extracting numbers from strings to obtain the Price value for each place
"""

df['Ticket Price in Thousands (IDR)'] = df['Tourism/Visitor Fee (approx in USD)'].str.extract('(\d+)', expand=False)
```

```python
df.head()
```

```python
df = df.drop(["Tourism/Visitor Fee (approx in USD)","Location","Description"], axis=1)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
"""
Replacing NaN values with a value of 0.
"""
df = df.fillna(0)
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
"""
Changing the data type of the "Ticket Price in Thousands (IDR)" column from object to integer.
"""

df['Ticket Price in Thousands (IDR)'] = df['Ticket Price in Thousands (IDR)'].astype(int)
```

**Finding Correlation**

```python
df_below_avg_rating["Google Maps Rating"].corr(df_below_avg_rating["Google Reviews (Count)"])
```

From the correlation values, it can be concluded that there is a correlation between Google Maps Rating and Google Reviews because they have a positive correlation value. However, this correlation is not strong as it is still far from the value of 1.

After identifying the correlation, the next step is to look for trends in the data using Scatter Plot.

```python
df_below_avg_rating.plot.scatter(x = 'Google Maps Rating', y = 'Google Reviews (Count)', s = 25);
```

Based on the correlation value of 0.4 and the scatter plot **showing no significant trend**, it can be concluded that **there is not a strong correlation between Google Maps Rating and Google Reviews**.
